# 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix. 

```python
fifa.dtypes
```

```python
fifa['Value']=fifa['Value'].replace({"€":" ","K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(float)
```

```python
fifa['Value'].dtypes
```

```python
fifa['Value']
```

```python
fifa['Wage']=fifa['Wage'].replace({"€":" ","K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(float)
```

```python
fifa['Wage'].dtypes
```

```python
fifa['Release Clause'].fillna(value = 0,inplace = True)
fifa['Release Clause']=fifa['Release Clause'].replace({"€":" ","K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(float)
```

```python
fifa['Release Clause'].dtypes
```

```python
fifa['Release Clause']
```

# 5. Convert the column "Joined" into integer data type with keeping only the year. 

```python
fifa["Joined"].fillna(value = 0,inplace = True)
fifa["Joined"]=pd.DatetimeIndex(fifa["Joined"]).year.astype(int)
```

```python
fifa["Joined"]
```

# 6. Convert the column "Contract Valid Until" to pandas datetime type.

```python
fifa["Contract Valid Until"].fillna(value = 0,inplace = True)
fifa["Contract Valid Until"]=pd.to_datetime(fifa["Contract Valid Until"])
```

```python
fifa["Contract Valid Until"]
```

# 7. The column 'Height' is in inches with a quotation mark, Convert to float with decimal points.

```python
fifa["Height"].fillna(value = 0.0,inplace = True)
fifa["Height"]=fifa["Height"].replace({"'":"."}, regex=True).map(pd.eval).astype(float)
```

```python
fifa["Height"]
```

# 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float.

```python
fifa["Weight"].fillna(value = 0.0,inplace = True)
fifa["Weight"]=fifa["Weight"].replace({"lbs":""} ,regex=True).map(pd.eval).astype(float)
```
